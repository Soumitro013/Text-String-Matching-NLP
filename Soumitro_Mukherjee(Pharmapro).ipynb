```python
Causes = data['List of Causes']
Causes
```

```python
Causes = Causes[:255]
```

```python
!pip install fuzzywuzzy
```

```python
Match = []
from fuzzywuzzy import fuzz
for i in Study:
  for j in Causes:
    Match.append(fuzz.partial_token_sort_ratio(i,j))
```

```python
results=np.array(Match).reshape(len(Study),len(Causes))
```

```python
Result_Frame=pd.DataFrame(data=results,columns=Causes, index=Study)
```
